```python
from postgres_connection import db
import pandas as pd

with db() as conn:
    cur = conn.execute('''
        SELECT * 
            FROM "EUGasNet"
            WHERE "import_country_code" = 'DE'
            LIMIT 10;
        ''')
    columns = [d[0] for d in cur.description]
    data = cur.fetchall()
        
df = pd.DataFrame(data, columns = columns)
df
```

```python
with db() as conn:
    cur = conn.execute('''
        SELECT COUNT(*) FROM "EUGasSC";
        SELECT COUNT(*) FROM "EUGasNet";
        ''')
    
    tables = ["EUGasSC", "EUGasNet"]
    for t in tables:
        record = cur.fetchone()
        print(f"Total rows in {t}: {record[0]}")
        if t != tables[-1]:    # Skip nextset() for last table.
            cur.nextset()
```
